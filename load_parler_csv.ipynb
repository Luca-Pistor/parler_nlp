```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('parlertrick/data/all_posts.csv')
```

```python
df
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from tqdm import tqdm_notebook as tqdm
tqdm().pandas() # ignore the annoying deprecation warnings that this throws
```

```python
posts = df['post_text'].tolist()
keep_tokens = []
for post in tqdm(posts):
    if isinstance(post, str):
        tokens = nltk.word_tokenize(post)
        keep = []
        for token in tokens:
            word = token.lower()
            if word.isalnum() and word not in nltk.corpus.stopwords.words('english'):
                keep.append(word)
        keep_tokens.append(keep)
    else:
        keep_tokens.append([])
```

```python
df['post_tokens'] = keep_tokens
```

```python
for idx,post in tqdm(enumerate(keep_tokens)):
    if post:
        if "inside" in post and "muriel" not in post:
            print(df.iloc[idx]['post_text'])
    else:
        pass
```

```python
import emoji
import regex

def isolate_emoji(post_text):

    emoji_list = []
    data = regex.findall(r'\X', post_text)
    for word in data:
        if any(char in emoji.UNICODE_EMOJI['en'] for char in word):
            emoji_list.append(word)

    return len(emoji_list)

def count_emoji(post_text):
    return len(isolate_emoji(post_text))
```

```python
df["post_text"] = df["post_text"].replace(np.nan, '', regex=True)
df.head()
```

```python
df["emoji_num"] = df["post_text"].apply(isolate_emoji)
df.head()
```
